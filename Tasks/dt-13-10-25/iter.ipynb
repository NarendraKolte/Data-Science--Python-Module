Dropwhile

```python
def greater_than_5(x):
    return x>5
```

```python
greater_than_5(25)
```

```python
from itertools import dropwhile
```

```python
list(dropwhile(greater_than_5,[1,2,5,7,1,5,2,4,25,30,60,70,80,90]))
```

```python
list(dropwhile(greater_than_5,[10,20,5,7,1,5,2,4]))
```
