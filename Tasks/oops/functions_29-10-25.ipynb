##Variable Length arg/parameter
tuple is given to function

```python
def var_fun(*x):
    print(x)
    print(type(x))
    sum=0
    for i in x:
        sum=i+sum
    return sum
```

```python
var_fun()
```

```python
var_fun(1,2,3,4)
```

```python
def v_fun(*x):
    print(x)
    print(type(x))
    sum=0
    for i in x:
        sum=i+sum
        yield sum
```

```python
v_fun(1,2,3,4,5)
```

```python
v1=v_fun(1,2,3,4,5)

list(v1)
```
