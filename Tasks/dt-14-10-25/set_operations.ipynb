Set Operations
1)UNION Combination of two strings

```python
a={1,2,3,4,5}
b={4,5,6,7,8}
```

```python
set(a).union(set(b))
```

2)Intersection

```python
a={45,78,96,54,32}
b={12,32,45,67,89}

print(a&b)
set(a).intersection(set(b))
```

3)Difference

```python
set(a).difference(set(b))
```

4)Symmetric Difference

```python
set(a).symmetric_difference(set(b))
```
