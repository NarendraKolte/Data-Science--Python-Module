```python
from abc import ABC
```

```python
class Shape(ABC):
    @abc.abstractmethod
    def area(self):
        pass

    def xyz(self):
        return "instance"

class Rectangle(Shape):
    def __init__(self, x, y):
        self.width = x
        self.height = y

    def area(self):
        print(self.width * self.height)
        return self.width * self.height
```

```python
r=Rectangle(10,20)
```

```python
print(r.xyz())
print(r.area())
```
